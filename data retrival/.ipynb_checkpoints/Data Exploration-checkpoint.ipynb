```python
import pandas as pd

df15 = pd.read_csv('final2015.csv')
df16 = pd.read_csv('final2016.csv')
df17 = pd.read_csv('final2017.csv')
df18 = pd.read_csv('final2018.csv')
df19 = pd.read_csv('final2019.csv')
```

```python
frames = [df15, df16, df17, df18, df19]
df = pd.concat(frames)
df.shape
```

```python
import numpy as np

np.random.seed(0)
```

```python
df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'], inplace=True)
```

```python
df.shape
```

```python
df['year'] = df['release_date'].map(lambda x: x[:4])
df.sample(30)
```

```python
dfmeans = df.groupby('year').mean()
```

```python
duplicated = df[df.duplicated(['track_name', 'release_date'])]
```

```python
df.drop_duplicates(subset=['artist_name','track_name'], inplace=True)
```
